```````````````````````````````python
import seaborn as sns
import pandas as pd

df = sns.load_dataset("titanic")
```````````````````````````````

```````````````````````````````python
df
```````````````````````````````

```````````````````````````````python
df.index
```````````````````````````````

```````````````````````````````python
df[0:13]
```````````````````````````````

```````````````````````````````python
delete_indexes = [0,2,4,6]
df.drop(delete_indexes,axis = 0).head() #inplace=True
```````````````````````````````

```````````````````````````````python
df.index = df.age
df.drop("age", axis=1,inplace=True)
df.head()
# index satirlarini yani soldaki
# 1234567 diye giden sayilar yerine yas kolonunu kullandik
```````````````````````````````

```````````````````````````````python
df.age = df.index
# tekrar ageyi ekledik

# indexide resetlemek istersek
df.reset_index(inplace=True)
df.head()
```````````````````````````````

```````````````````````````````python
"age" in df
```````````````````````````````

```````````````````````````````python
print(df["age"].head())
print(type(df["age"].head()))
# pandas serisi olarak geldi
```````````````````````````````

```````````````````````````````python
print(df[["age"]].head())
print(type(df[["age"]].head()))
# pandas dataframe olarak geldi
```````````````````````````````

```````````````````````````````python
# birden fazla col secimi
#1
df[["age", "alive"]]
#2
col_names = ["age", "alive"]
df[col_names]
```````````````````````````````

```````````````````````````````python
df["age_new"] = df["age"] ** 2
df.head()
# age sutunun karesini alir ve yeni colon olusturur
```````````````````````````````

```````````````````````````````python
# kolonlardaki belirli str ifadelerini secme

print(df.loc[:, df.columns.str.contains("age")].head())
# kolon isimlerinde age ifadesini barindiranlari secti
print(df.loc[:, ~df.columns.str.contains("age")].head())
# ~ tilda isareti onun disindakilerini sec anlamina getirdi
```````````````````````````````

```````````````````````````````python
# loc & iloc

# loc label based selection Satır ve sütun isimlerini kullanarak veri seçimi yaparsınız. İndeksler ve sütun isimleri ile çalışır.
# iloc integer based selection Satır ve sütunların pozisyon numaralarını kullanarak veri seçimi yaparsınız. Yani, indekslerin sıralı numaraları ile çalışır.

print(df.loc[0:3]) # 0dan 3e kadar olan idexleri getirir
print(df.iloc[0,3]) # 0.satir 3.sutundaki elemani getirir

print(df.iloc[0:3, 0:1])
print(df.loc[0:3, "age"])

col_names = ["age", "embarked", "alive"]
print(df.loc[0:3, col_names])
```````````````````````````````

```````````````````````````````python
# kosullu islemler

#value_counts frekanslarini verir
print(df['embark_town'].value_counts())

print(df[df["age"] > 50].head())
print(df[df["age"] > 50]["age"].count())
#degisken ismi girilecekse loc
print(df.loc[(df["age"] > 50) & (df["sex"] == "male"), ["age","class"]].head())
```````````````````````````````

```````````````````````````````python
# Toplulastirma ve Gruplama

# toplulastirma ozet istatistik veren fonksiyonlardir (first last mean median min max std count)
# gruplastirma agg ile yapilandir

print(df["age"].mean())


print(df.groupby("sex")["age"].mean())
# bu sekilde degilde

print(df.groupby("sex").agg({"age": ["mean","sum","min","max","count"]}))
# bu sekilde uygula cunku hem mean hem sum ne istersen ekleyebilirsin

print(df.groupby("sex").agg({"age": ["mean","sum","min","max","count"],
                             "survived": "mean"}))


print(df.groupby(["class","embark_town"]).agg({"age":"mean",
                             "survived":"mean"}))

print(df.groupby(["sex", "class"]).agg({
    "age":"mean",
    "survived":"mean",
    "sex":"count"
}))
```````````````````````````````

```````````````````````````````python
# Pivot Table

print(df.pivot_table("survived", "sex", "embarked"))
print("``````````````````````````````")
print(df.pivot_table("survived", "sex", ["embarked", "class"]))
print("``````````````````````````````")
print(df.pivot_table("survived", "sex", ["embarked", "class"], aggfunc="mean"))
```````````````````````````````

```````````````````````````````python
# deger siniflandirma yapabilmemizi saglar
df["new_age"] = pd.cut(df["age"], [0,10,18,25,40,90])
df.pivot_table("survived", "sex", "new_age")

# 0,10 arasindaki yaslardan kadinlarin %60 i hayatta kalmis erkeklerin ise %57 si
```````````````````````````````

```````````````````````````````python
pd.set_option("display.width", 200) # ekrana sigmasini saglar
df.pivot_table("survived", "sex", ["new_age", "class"])
```````````````````````````````

```````````````````````````````python
# Apply & Lambda

df["age2"] = df["age"] * 2
df["age3"] = df["age"] * 5

# 10'a bolmek icin

# df["age2"] / 10 yerine

for col in df.columns.values:
    if "age" in col:
        df[col] = df[col] / 10
        print(df[col].head())

# ya da

df[["age", "age2", "age3"]].apply(lambda x: x / 10).head()

# daha sistematik sekilde
df.loc[:, df.columns.str.contains("age")].apply(lambda x: x / 10).head()


```````````````````````````````
