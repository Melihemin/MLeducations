```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 100)
df = sns.load_dataset('titanic')
df.head()
```

# Matplotlib

* Sayisal degisken: (hist, boxplot)

## Matplotlib Ozellikleri

```python
# plot

x = np.array([1, 100])
y = np.array([0, 150])

plt.plot(x,y, color="green")

plt.plot(x, y, 'o', color="orange")
plt.show()
```

```python
y = np.array([1,10,15,2,4,5,6,7,7,19,5])

plt.plot(y,color= "orange" ,marker="o")
plt.show()
```

```python
# Line

y = np.array([1,10,15,2,4,5,6,7,7,19,5])

plt.plot(y,color= "orange" ,linestyle="--") # dashdot da yazilabilir
plt.show()
```

```python
x = np.random.randint(1,100,5)
y = np.random.randint(50,100,5)

plt.plot(x, color="green", marker="o")
plt.plot(y, color="red", linestyle="--")
plt.grid()
plt.show()
```

```python
# Labels

x = np.random.randint(1,100,5)
y = np.random.randint(1,50,5)

plt.title("Test")
plt.xlabel("X")
plt.ylabel("Y")


plt.plot(x, color="green", marker="o")
plt.plot(y, color="red", linestyle="--")
plt.show()
```

```python
# Subplots

x = np.random.randint(1,100,5)
y = np.random.randint(1,50,5)

plt.subplot(1,2,1)
plt.plot(x, color="green", marker="o")

plt.subplot(1,2,2)
plt.plot(y, color="red", linestyle="--")
plt.show()
```

```python
# Aylik harcama grafigim
price = np.array([230, 540, 275, 900, 400, 245, 540,550, 350, 200, 450, 970])
days = np.arange(len(price)) # gun sayisini ayarla
gunluk = []
for i in range(1,12):
  gunluk.append(350)

gunluk_arr = np.array(gunluk)
plt.plot(days, price, marker="o", color="red") # harcamani goster
plt.plot(gunluk_arr, color="green", linestyle="--") # gunluk harcama tutarini belirt
plt.xlabel("Gunler") # x ekseni
plt.ylabel("Harcanan Tutar") # y ekseni
plt.title("Aylık Harcama(Veri kaynağı ziraat mobil)[Gunluk harcanmasi gereken tutar yesil cizgi ile gosterilmistir]") # Add a title
plt.grid() # verilerin daha iyi okunmasi icin grid ekle
plt.show() # plotu goster
```
