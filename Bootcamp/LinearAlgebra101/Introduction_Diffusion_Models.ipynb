```python
from diffusers import StableDiffusionPipeline
import torch

# Modeli yükle
model_id = "CompVis/stable-diffusion-v1-4-original"
pipe = StableDiffusionPipeline.from_pretrained(model_id)
pipe.to("cpu")  # Eğer GPU kullanıyorsanız 'cuda' kullanın, aksi halde 'cpu'

# Metin girişi
prompt = "a fantasy landscape, mountains, river, and a castle in the distance"

# Görüntü oluşturma
with torch.no_grad():
    image = pipe(prompt).images[0]  # Metinden ilk görüntüyü al

# Görüntüyü göster
import matplotlib.pyplot as plt

plt.imshow(image)
plt.axis("off")
plt.show()
```
