```python
import numpy as np
```

```python
# Katsayilar matrisi A

A = np.array([[2,3,-1],
              [4,1,2],
              [-1, 2, 3]])

A.shape
```

```python
# Sabit terimler vektoru

b = np.array([1, -2, 3])
```

```python
A_ters = np.linalg.inv(A)
A_ters
```

```python
X = np.matmul(A_ters, b)
X
```

```python
print("X1", X[0])
print("X2", X[1])
print("X3", X[2])
```

```python
# Fonksiyon hali

def cozucu(Katsayilar,SabitTerimler):

  try:

    a = np.array(Katsayilar)
    b = np.array(SabitTerimler)

    a_ters = np.linalg.inv(a)
    x = np.matmul(a_ters, b)

    print("X1", x[0])
    print("X2", x[1])
    print("X3", x[2])

  except np.linalg.LinAlgError:
    print("Sistemin cozumu yoktur.")
```

```python
cozucu(A,b)
```

```python
x = np.linalg.solve(A,b)
x
```
